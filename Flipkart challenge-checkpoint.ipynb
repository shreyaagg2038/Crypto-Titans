```python
df=df.replace('\*|\[|\]|','',regex=True).astype(str)
df
```

```python
newdf=(df.set_index(df.columns.drop('item',1).tolist())
   ...:    .item.str.split(',', expand=True)
   ...:    .stack()
   ...:    .reset_index()
   ...:    .rename(columns={0:'item'})
   ...:    .loc[:, df.columns]
   ...: )
newdf
```

```python
newdf['item'].str.replace("'","")
newdf
```

```python
count_products = pd.DataFrame(newdf.groupby('item')['orderid'].count())
count_products = count_products.sort_values('orderid', ascending=False)
count_products
```

```python
count_products.rename(columns = {'orderid':'count'}, inplace = True)
count_products
                                 
```

```python
C = count_products['count'].sum()
print(C)
```

```python
def rating(x, C=C):
    v = x['count']
    return v*100/C
```

```python
count_products['rating_calculated']=count_products.apply(rating, axis=1)
count_products
```

```python
merged_inner = pd.merge(left=newdf, right=count_products, left_on='item', right_on='item')
```

```python
merged_inner
```

```python
ratings_utility_matrix = merged_inner.pivot_table(values='rating_calculated', index='customerid', columns='item', fill_value=0)
ratings_utility_matrix.head()
```

```python
X = ratings_utility_matrix.T
X.head()
```

```python
X1 = X
```

```python
import sklearn
from sklearn.decomposition import TruncatedSVD
```

```python
SVD = TruncatedSVD(n_components=10)
decomposed_matrix = SVD.fit_transform(X)
decomposed_matrix.shape
```

```python
correlation_matrix = np.corrcoef(decomposed_matrix)
correlation_matrix.shape
```
